<b>Getting Data from internet page and using Soup to parse
    </b>

```python
# getting data from internet
source='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page= requests.get(source).text
```

<B>Converting to readable Table </b>

```python
soup = BeautifulSoup(page)
print(soup.title)
tabledemo=str(soup.table)

display_html(tabledemo,raw=True)
```

```python

dfs = pd.read_html(tabledemo)
df=dfs[0]
df.head()
```

<b>Removing 'Not Assigned' from Borough
</b>

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

#df1
```

<b>Grouping and Joining..adding Commas and reset index</b>

```python
# Combining the neighborhoods with same Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)
df2
```

<b>Changing / to ,</b>

```python
df2['Neighborhood']=df2['Neighborhood'].replace('Regent Park / Harbourfront', 'Regent Park, Harbourfront')

df2['Neighborhood']=df2['Neighborhood'].replace('Lawrence Manor / Lawrence Heights', 'Lawrence Manor, Lawrence Heights')
df2['Neighborhood']=df2['Neighborhood'].replace('Queen\'s Park / Ontario Provincial Government', 'Queen\'s Park, Ontario Provincial Government')
df2['Neighborhood']=df2['Neighborhood'].replace('Malvern / Rouge', 'Malvern, Rouge')
df2['Neighborhood']=df2['Neighborhood'].replace('Parkview Hill / Woodbine Gardens', 'Parkview Hill, Woodbine Gardens')
df2['Neighborhood']=df2['Neighborhood'].replace('Garden District / Ryerson', 'Garden District, Ryerson')
df2['Neighborhood']=df2['Neighborhood'].replace('West Deane Park / Princess Gardens / Martin Grove / Islington / Cloverdale', 'West Deane Park, Princess Gardens, Martin Grove, Islington, Cloverdale')
df2['Neighborhood']=df2['Neighborhood'].replace('Rouge Hill / Port Union / Highland Creek', 'Rouge Hill, Port Union, Highland Creek')
df2['Neighborhood']=df2['Neighborhood'].replace('Guildwood / Morningside / West Hill', 'Guildwood, Morningside, West Hill')
df2['Neighborhood']=df2['Neighborhood'].replace('Eringate / Bloordale Gardens / Old Burnhamthorpe / Markland Wood', 'Eringate, Bloordale Gardens, Old Burnhamthorpe, Markland Wood')
df2['Neighborhood']=df2['Neighborhood'].replace('Bathurst Manor / Wilson Heights / Downsview North', 'Bathurst Manor, Wilson Heights, Downsview North')
df2['Neighborhood']=df2['Neighborhood'].replace('Richmond / Adelaide / King', 'Richmond, Adelaide, King')
df2['Neighborhood']=df2['Neighborhood'].replace('Old Mill South / King\'s Mill Park / Sunnylea / Humber Bay / Mimico NE / The Queensway East / Royal York South East / Kingsway Park South East', 'Old Mill South, King\'s Mill Park, Sunnylea, Humber Bay, Mimico NE, The Queensway East, Royal York South East, Kingsway Park South East')
df2['Neighborhood']=df2['Neighborhood'].replace('Mimico NW / The Queensway West / South of Bloor / Kingsway Park South West / Royal York South West', 'Mimico NW, The Queensway West, South of Bloor, Kingsway Park South West, Royal York South West')
df2['Neighborhood']=df2['Neighborhood'].replace('The Kingsway / Montgomery Road / Old Mill North', 'The Kingsway, Montgomery Road, Old Mill North')
```
